```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,3,50)
display(x)
figure = plt.figure(figsize=(8,4)) #設定底圖
y=x**2


axes = figure.add_subplot() #設定圖表
axes.plot(x) #設定藍線
axes.plot(y) #設定橘線
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
t = np.arange(0,1,0.5)
display(t)

y1 = np.sin(2*np.pi * t)
y2 = np.cos(2 * np.pi * t)
display(y1)
display(y2)
figure1=plt.figure(figsize=(8,8))

axes1 = figure1.add_subplot(2,2,1) 
axes1.plot(y1) 
axes1.plot(y2)

r = np.arange(-1,1,0.1)
display(r)
C=2*np.pi*r
#display(np.pi)
display(C)
axes1 = figure1.add_subplot(2,2,1) 
axes1.plot([r,C]) 


plt.show()
```
